```python
import pandas as p
import numpy as n
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import load_breast_cancer as b
d=b()
d
```

```python
d.keys()
```

```python
r=d.data
c=d.feature_names
df=p.DataFrame(r,columns=c)
df["Target"]=d.target
df.sample(5)
```

```python
df.isnull()
```

```python
df.info()
```

```python
x=df.drop("Target",axis=1)
y=df["Target"]
```

```python
from sklearn.model_selection import train_test_split as tt
x_train,x_test,y_train,y_test=tt(x,y,test_size=0.3,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression as l
obj=l()
obj.fit(x_train,y_train)
obj.score(x_train,y_train)*100
```

```python
#obj.predict([[10.97,17.20,71.73,371.5,0.08915,0.11130,0.09457,0.036130,0.1489,0.06640,26.87,90.14,476.4,0.13910,0.4082,0.4779,0.15550,0.2540,0.09532]])
yp=obj.predict(x_test)

s=p.DataFrame({'a':y_test,'p':yp})
s.sample(10)
```
