```python
### support vector machine:
## it is supervised ML algorithm
## it is used for both classification and regression problems
## it provides maximum distance(or) boundaries between classes using hyperplane
## boundary line is also known as hyperplane
## it use the another hyperplanes for classifing data using support vectors.
```

```python
import numpy as n
import pandas as p
```

```python
# read .csv files into dataframe
df=p.read_csv("mushrooms.csv")
df.sample(5)
```

```python
df.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder as le
l=le()
df["class"]=l.fit_transform(df["class"])
df.head()
```

```python
df.columns
```

```python
for col in df.columns:
#print(col)
    
    df[col]=l.fit_transform(df[col])
    df.head()
```

```python
# cake input and output
x=df.drop("class",axis=1)
y=df["class"]
```

```python
from sklearn.model_selection import train_test_split as tt
x_train,x_test,y_train,y_test=tt(x,y,test_size=0.3,random_state=42)
```

```python
from sklearn.svm import SVC as sv
model=sv()
model.fit(x_train,y_train)
model.score(x_train,y_train)
```

```python
# predict the test data
ypre=model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score as a
a(y_test,ypre)*100
```

```python
from sklearn.metrics import classification_report as c
c(y_test,ypre)*100
print(c(y_test,ypre))

```

```python
from sklearn.metrics import confusion_matrix as cm
cm(y_test,ypre)
```
