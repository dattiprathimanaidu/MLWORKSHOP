```python
import numpy as n
import pandas as p
```

```python
from sklearn.datasets import load_diabetes as l
d=l()
d
d.keys()
```

```python
r=d.data
c=d.feature_names
df=p.DataFrame(r,columns=c)
df["Target"]=d.target
df.head()
```

```python
df.corr()
```

```python
x=df[["bmi","s4","s5"]]
y=df["Target"]

```

```python
x=df.drop("Target",axis=1)
x.head()
```

```python
from sklearn.neighbors import KNeighborsClassifier as l
obj=l(n_neighbors=4)
obj
```

```python
obj.fit(x,y)
```

```python
obj.predict(x)
```

```python
from sklearn.metrics import accuracy_score
m=obj.predict(x)
```

```python
accuracy_score(y,m)*100
```
