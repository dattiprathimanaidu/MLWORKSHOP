```python
import pandas as p
import numpy as n
```

```python
data=n.array([[10,n.nan,11,15,n.nan],[12,13,14,20,25],[13,89,90,23,24],[n.nan,23,n.nan,56,78],[12,n.nan,34,23,n.nan]])
data
```

```python
df=p.DataFrame(data,columns=["a","b","c","d","e"])
df
```

```python
df.isnull().sum()
```

```python
df.dropna()
```

```python
df["a"]=df["a"].fillna(df["a"].mean())
df
```

```python
df["c"]=df["c"].fillna(df["c"].median())
df
```

```python
df["b"]=df["b"].fillna(df["b"].mean())
df
```

```python
df["e"]=df["e"].fillna(method="ffill")
df
```

```python
df["e"]=df["e"].fillna(method="bfill")
df
```

```python
data=n.array([[10,n.nan,11,15,n.nan],[12,13,14,20,25],[13,89,90,23,24],[n.nan,23,n.nan,56,78],[12,n.nan,34,23,n.nan]])

df=p.DataFrame(data,columns=["a","b","c","d","e"])
df
```

```python
from sklearn.impute import SimpleImputer as im
```

```python
si=im(strategy="median")
obj3=si.fit_transform(df)
obj3
```

```python
df2=p.DataFrame(obj3,columns=df.columns)
df2
```

```python
si=im(strategy="mean")
obj3=si.fit_transform(df)
obj3
```

```python
df3=p.DataFrame(obj3,columns=df.columns)
df3
```

```python
si=im(strategy="constant")
obj3=si.fit_transform(df)
obj3
```

```python
df4=p.DataFrame(obj3,columns=df.columns)
df4
```
